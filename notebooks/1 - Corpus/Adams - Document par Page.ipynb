# Page dans adams

```python
import lxml.etree as ET

NS = {"t": "http://www.tei-c.org/ns/1.0"}

xml = ET.parse("../../these-corpus/collated.xml")
pages = [
    int(page)
    for page in xml.xpath("//t:biblScope[@unit='page']/text()", namespaces=NS)
]
        
```

```python
from collections import Counter
import matplotlib.pyplot as plt
from pandas import Series

fig, ax = plt.subplots(figsize=(7,5), dpi=300)

fig.patch.set_facecolor('white')

s = Series(Counter(sorted(pages)))
s.plot(kind="area", rot=True, ax=ax, grid=True, 
       xlabel="Reference Page", ylabel="Number of cited passages")

ax.set_title(
    f"Evolution of the number of passages found in \nAdams' Latin Sexual Vocabulary (total = {len(pages)})"
)
fig.show()
```
