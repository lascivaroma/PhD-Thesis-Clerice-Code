```python
import glob
from MyCapytain.resolvers.cts.local import CtsCapitainsLocalResolver

resolver = CtsCapitainsLocalResolver(
    ["../data/raw/corpora/PerseusDL_canonical-latinLit/PerseusDL_canonical-latinLit"]+ \
    ["../data/raw/corpora/OpenGreekAndLatin_csel-dev/OpenGreekAndLatin_csel-dev"]
)
```

```python
from MyCapytain.common.constants import Mimetypes

xpath_citations = []
citations = []

for readable in resolver.getMetadata().readableDescendants:
    if readable.citation:
        xpath_citations.append(readable.citation.export(Mimetypes.XML.TEI))
        citations.append("|".join([cite.name or "" for cite in readable.citation]))
```

```python
from tabulate import tabulate

rows = [
    ["", "Unique", "Total"],
    ["XML Citation Structures", len(set(xpath_citations)), len(xpath_citations)],
    ["Citation labels", len(set(citations)), len(xpath_citations)]
]
print(tabulate(rows,  tablefmt="github"))
```
